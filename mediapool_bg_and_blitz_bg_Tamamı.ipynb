**BLİTZ.BG**

```python
def datecreator():
  links = []
  link = "https://blitz.bg/analizi-i-komentari/alfagari-s-tezhki-dumi-za-bolestnoto-sstoyanie-na-slavi_news" 
  for i in range(5310,5315,1):
    url = "{}{}{}".format(link,i,'.html')
    links.append(url)
    with open("blitz.bg_links.txt", 'a') as file: 
        file.write(url+'\n')
  return links
```

```python
links = datecreator()
print(links)
```

```python
def creator(url):
    r = requests.get(url)   
    soup = BeautifulSoup(r.content, 'html5lib')
    title = soup.find("h1").getText()
    content_array = soup.find("div", attrs = {"class":"article-text"}).getText()
    # category = soup.find("span", attrs = {"itemprop":"name"}).getText()
    # print(category)
    date = soup.find("div", attrs = {"class":"article-info"}).getText()
    date = date.split()
    date = date[2] + date[3] + date[4]
    content_array = content_array.split()
    content_string = ""
    stop = "/*"
    for w in content_array:
      if stop == w:
        break
      else:
        content_string = content_string+" "+w
    w_data="{};{};{};{}".format(url,date,title,content_string)
    with open("blitz.bg_content.txt", 'a') as file: 
        file.write(w_data+'\n')
```

```python
# creator('https://blitz.bg/analizi-i-komentari/alfagari-s-tezhki-dumi-za-bolestnoto-sstoyanie-na-slavi_news5313.html')
```

```python
links=[]
with open("blitz.bg_links.txt",'r',newline='') as f:
  for i in f.readlines():    
    links.append(i)
```

```python
t1=time.time()
with concurrent.futures.ProcessPoolExecutor() as execut:
  b_res=[execut.submit(creator,i.strip()) for i in links]

print(time.time()-t1)
```

**MEDİAPOOL.BG**

```python
def datecreator():
  links = []
  link = "https://www.mediapool.bg/bnb-razreshi-na-evrobank-da-kupi-iztsyalo-investitsionnoto-druzhestvo-transinvestmant-news" 
  for i in range(100000,100010,1):
    url = "{}{}{}".format(link,i,'.html')
    links.append(url)
    with open("mediapool.bg_links.txt", 'a') as file: 
        file.write(url+'\n')
  return links
```

```python
links = datecreator()
# print(links)
```

```python
def creator(url):
    r = requests.get(url)   
    soup = BeautifulSoup(r.content, 'html5lib')
    title = soup.find("h1").getText()
    content_array = soup.find("div", attrs = {"id":"art_font_size"}).getText()
    category = soup.find("a", attrs = {"itemprop":"articleSection"}).getText()
    date = soup.find("time").getText()
    date = date.split()
    date = date[2] +':' +date[3] +':'+ date[4]
    content_array = content_array.split()
    content_string = ""
    stop = "/*"
    for w in content_array:
      if stop == w:
        break
      else:
        content_string = content_string+" "+w
    w_data="{};{};{};{};{}".format(url,date,category,title,content_string)
    # print(w_data)
    with open("mediapool.bg_content.txt", 'a') as file: 
        file.write(w_data+'\n')
```
