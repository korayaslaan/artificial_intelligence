```python
def datecreator():
  links = []
  t1=datetime(2019,4,9)#en son 1999 05 01 e kadar çalışıyo
  t2=datetime(2019,4,10)
  t = timedelta(days = 1)
  dates = np.arange(t1, t2, t).astype(datetime)
  link ="https://www.hurriyet.com.tr/arama/#/?page=" #1000 e kadar sayfa var kanser
  for date in dates:
    newdate=date.strftime('%Y/%m/%d')
    enddate = pd.to_datetime(newdate) + pd.DateOffset(days=1)
    enddate =enddate.strftime('%Y/%m/%d')
    links.append("{}{}{}{}{}{}{}".format(link,'1','&where=hurriyet&how=Article&startDate=',newdate,'&finishDate=',enddate,'&platform=hurriyet&isDetail=true'))
  return links
```

```python
def page_count_finder():
  date = datecreator()
  for i in date:
    browser=wd
    browser.get(i)   
    page_sour=browser.page_source
    soup = BeautifulSoup(page_sour, 'html5lib')
    z=soup.find("span",{"class":"totalArticle"})
    h=(float(z.getText()))/10
    return (math.ceil(h)+1) # ceil tavan 15.1 =16    round taban => 15.9 =15
```

```python
print(page_count_finder())
```

```python
def get_link(i):
    sayfa =i
    browser = wd
    browser.get(sayfa)
    test = []
    count = 0
    count2 = 0
    count3 = 0
    dizi = []
    sayfa_kaynağı = browser.page_source
    soup = BeautifulSoup(sayfa_kaynağı, "html.parser")
    for a in soup.find_all('a', href=True):
        count2+=1
        link=a['href']
        test.append(link)
    for i in test:
      result = i.startswith("http://www.hurriyet.com.tr")
      if result == True:
        if count> 15 and count%5 == 2 and len(i)>50 and count3<10 :#
          with open("hürriyet_link_tamamı.txt", 'a') as file: 
             file.write(i+'\n')
             dizi.append(i)
          count3+=1
      count+=1
    browser.refresh()
    return dizi
```

```python
def creator(url):
    r = requests.get(url)   
    soup = BeautifulSoup(r.content, 'html5lib')
    title = soup.find("h1", attrs = {"class":"rhd-article-title"}).getText()
    category = soup.find("div", attrs = {"class":"rhd-category-box"}).getText()
    content_array = soup.find("div", attrs = {"class":"rhd-all-article-detail"}).getText()
    date = soup.find("span", attrs = {"class":"rhd-time-box-text hidden-sm-down"}).getText()
    date = date.split()
    date = date[2]
    content_array = content_array.split()
    content_string = ""
    stop = "(function(url,"
    stop2 = "namespace)"
    for w in content_array:
      if w == stop or w==stop2:
        break
      else:
        content_string = content_string+" "+w
        #w_data = url+";"+date+";"+title+";"+content_string
    w_data="{};{};{};{};{}".format(url,date,category,title,content_string)
    print(w_data)
    # write_to_txt(w_data)
    with open("hürriyet_content.txt", 'a') as file: 
        file.write(w_data+'\n')
```

```python
links = datecreator()

t1=time.time()
with concurrent.futures.ProcessPoolExecutor() as execut:
  b_res=[execut.submit(get_link,i.strip()) for i in links]

print(time.time()-t1)
```
