```python
def datecreator():
  date_list=[]
  txt=""
  newdate=""
  url_in=""
  url1="https://www.sabah.com.tr/get/arama?query=kanser&page="
  url2="https://www.sabah.com.tr/get/arama?query=prostat%20kanseri&page="
  url3="https://www.sabah.com.tr/get/arama?query=meme%20kanseri&page="
 
 
  for i in range(2): #3050 sayfa var
      url_in="{}{}".format(url1,i+1)
      date_list.append(url_in)
  for i in range(2): #2210 sayfa var
      url_in="{}{}".format(url2,i+1)
      date_list.append(url_in)
  for i in range(2): #2450 sayfa var
      url_in="{}{}".format(url3,i+1)
      date_list.append(url_in)


  return date_list
```

```python
def get_link(link):
  r = requests.get(link)   
  soup = BeautifulSoup(r.content, 'html5lib')
  url = 'https://www.sabah.com.tr'
  dizi = []
  for a in soup.find_all('a', href=True):
    a = a['href']
    result = a.startswith('/')
    if result == True:
      url_in="{}{}".format(url,a)
      dizi.append(url_in)
  return dizi    
```

```python
def write_to_txt(data):
  with open("sabahkanser.txt", 'a') as file: 
    file.write(data+'\n')
    file.truncate()
```

```python
def veri_cek(url):
  print(url)
  r = requests.get(url)  
  soup = BeautifulSoup(r.content, 'html5lib')
  title = soup.find("h1", attrs = {"class":"pageTitle"}).getText()
  content_array = soup.find("div", attrs = {"class":"newsBox"}).getText()
  date = soup.find("span", attrs = {"class":"textInfo"}).getText()
  date = date.split()
  date = date[2]
  content_array = content_array.split()
  content_string = ""
  for w in content_array:
    content_string = content_string+" "+w
  #w_data = url+";"+date+";"+title+";"+content_string
  data="{};{};{};{}".format(url,date,title,content_string)
  print(data)
  # write_to_txt(data)
  with open("sabahkanser.txt", 'a') as file: 
       file.write(data+'\n')
```

```python
veri_cek()
```

```python
def get_url(url):
    dizi = url
    dizi2 = []
    for i in dizi:
         if not i in dizi2:           #çekilen linklerde aynı olan verileri teke düşürüyor 
             dizi2.append(i)
    return dizi2
```

```python
date = datecreator()
dizi = []
dizi2 = []
for i in date:
  dizi = dizi + get_link(i)
dizi2 = get_url(dizi)
print(len(dizi2))
for i in dizi2:
  try:
    veri_cek(i)
  except: 
    print("Başarısız Link")
```
